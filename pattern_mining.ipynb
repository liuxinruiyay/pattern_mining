```python
import matplotlib.pyplot as plt
import pandas as pd
import tqdm
from efficient_apriori import apriori

data = pd.read_csv("./data/winemag-data_first150k.csv")
data.head()
```

#对数据集进行预处理，并主要分析葡萄酒生产国家，生产省份，葡萄酒品种和打分之间的关系

```python
print("原始数据集行数：",len(data))
data = data[['country','province','variety','points']]
data = data.dropna(how='any')
print("去掉缺失值后的数据为",len(data))
```

组合country和province生成新属性location

```python
data['location'] = data['country'] + '-' + data['province']
data = data[['location', 'variety', 'points']]
data.head()
```

查看连续性属性points的直方图属性，发现数据分布在[80,100]之间

```python
data['points'].plot.hist(bins=21)
```

```python
values = [75, 80, 85,90,95,100]
data['points'] = pd.cut(data['points'],values)
data['points'] = data['points'].astype('str')
data['points'].value_counts()
```

```python
data.head()
```

```python
apriori_data = []
for _, (loc, var, poi) in tqdm.tqdm(data.iterrows(), total=len(data)):
    apriori_data.append((loc, var, poi))
```

找出频繁模式，并展示频繁模式

```python
itemsets, rules = apriori(apriori_data, min_confidence=0.3, min_support=0.005)
itemsets
```

展示关联规则并计算其支持度和置信度

```python
rules
```

```python
for rule in sorted(rules, key=lambda rule: rule.confidence):
    print(repr(rule), "support:", rule.support, "confidence:", rule.confidence)
```

使用其他评价指标对规则进行评估：lift和卡方

```python
for rule in sorted(rules, key=lambda rule:rule.confidence):
    print(repr(rule),"lift",rule.lift, "conviction:",rule.conviction)
```

对挖掘结果进行可视化展示

```python
def visual(rule_list):
    rules = []
    conf = []
    support = []
    lift = []
    for rule in rule_list:
        rules.append(repr(rule))
        conf.append(rule.confidence)
        support.append(rule.support)
        lift.append(rule.lift)
    plot_bar(rules, conf, "rules-confidence")
    plot_bar(rules, support, "rules-support")
    plot_bar(rules, lift, "rules-lift")

def plot_bar(rules, data, title):
    plt.title(title)
    plt.xticks(range(len(data)), rules, rotation = 90)
    plt.bar(range(len(data)), data)
    plt.show()
visual(sorted(rules, key=lambda rule:rule.confidence)[:20])
```
